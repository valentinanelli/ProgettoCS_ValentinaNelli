## Punto 12
#### For each year, compute the total amount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st, 2016, planned expiration time January 30th 2018, and amount 5000USD has an amount of 5000USD * 31 / (31+365+30) = 363.85 for 2016, 5000USD * 365 / (31+365+30) = 4284.04 for 2017, and 5000USD * 30 / (31+365+30) = 352.11 for 2018.

```python
#Creo una funzione che restituisce il numero di giorni che intercorrono tra due date
def day_calc(date1, date2):
    
    #in alcuni casi planned_expiration_time è più lungo rispetto a disburse time,
    #Quindi imposto la funzione facendo in modo che la data 1 sia quella inferiore
    (begin,end) = (date1,date2) if date1<date2 else (date2,date1)
    date_list = list()
    date_list.append(begin)

    #creo una lista che parte da begin e finsce ad end e parte dal 1 gennaio dell'anno y
    for y in range(begin.year+1, end.year+1):
        date_list.append(pd.Timestamp(year=y, month=1, day=1, tz='utc'))
    date_list.append(end)
    
    #creo un dizionario che mi dice per ogni anno quanti giorni sono passati tra disburse time e planned_expiration_time
    res = dict()
    for i in range(len(date_list)-1):
        res[date_list[i].year] = (date_list[i+1] - date_list[i]).days

    return pd.Series(res, name = 'DateValue')

data_loans = df_loans.dropna(subset = ['planned_expiration_time', 'disburse_time'])
```

```python
# applico la funzione day_calc usando .apply 
df_day = data_loans.apply(lambda x: day_calc(x['planned_expiration_time'],x['disburse_time']), axis = 1)
```

```python
# alcune durate sono in valore negativo, quindi le trasformo in positivo 
df_loans['duration'] = df_loans['duration'].abs()

#unisco il dataset precedente all'ammontare di soldi prestati e la durata 
df_day.fillna(0, inplace= True)
df_day = pd.concat([df_day, df_loans['loan_amount']
                             , df_loans['duration'] ]
                            , axis = 1, sort = True)
```

```python
#divido tutte le date per la duration e moltiplico per amount
loan = df_day[[2011,2012,2013,2014,2015,2016,2017,2018
                        ]].div(df_day.duration, axis=0).mul(df_day.loan_amount, axis = 0)

# Importo totale dei prestiti per ogni anno
loan.sum()
```
